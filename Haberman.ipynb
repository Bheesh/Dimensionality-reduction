```python
##Importing csv file for initial data analysis
## Importing packages which helps in numerical interpretation

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

haber=pd.read_csv("C:/Users/MASTER/Desktop/haberman.csv")
haber['status']=haber['status'].map({1:'yes',2:'no'})
haber['status']=haber['status'].astype('category')
print(haber.shape)
print(haber.columns)
```

```python
##Finding mean of age , year of treatment and auxillarry nodes
mean=np.mean(haber.age)
print("mean of age",mean)
mean=np.mean(haber.year)
print("mean of year of treatment",mean)
mean=np.mean(haber.node)
print("mean of auxillary nodes",mean)
```

```python
## Using 2-D scatter plot 

haber.plot(kind="scatter",x="year",y="node")
plt.xlabel("year")
plt.ylabel("node")
plt.title("haber")
plt.legend("upper left")
plt.show()
```

```python
##Information from 1-D plot is not agreeable
##2-D plots are used with seaborn package
## Difference can n be oberved as colours are different
sns.set_style("whitegrid");
sns.FacetGrid(haber, hue="status", size=6) \
   .map(plt.scatter, "year", "node") \
   .add_legend();
plt.show();
```

```python
## Plots with given parameters
sns.FacetGrid(haber, hue="status", size=5) \
   .map(sns.distplot, "age") \
   .add_legend();
plt.show();

sns.FacetGrid(haber, hue="status", size=5) \
   .map(sns.distplot, "year") \
   .add_legend();
plt.show();

sns.FacetGrid(haber, hue="status", size=5) \
   .map(sns.distplot, "node") \
   .add_legend();
plt.show();
```

```python
## PDF and CDF
##PDF helps in finding distribution in random variables

counts, bin_edges = np.histogram(haber['age'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
print("PDF is",pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
print("CDF is",cdf);
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:],cdf);
plt.title("Age of patient")
plt.show();

## PDF and CDF of year of treatment
counts, bin_edges = np.histogram(haber['year'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
print(cdf);
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:],cdf);
plt.title("Year of treatment ")
plt.show();

## PDF and CDF of auxillary nodes
counts, bin_edges = np.histogram(haber['node'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
print(cdf);
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:],cdf);
plt.title("Node ")
plt.show();
```

```python
##Using pairplots for better understanding 
##Much clarity will be provided as plots are taken against all parameters 

plt.close()
sns.set_style("whitegrid");
sns.pairplot(haber,vars=['age','year','node'],hue='status',size=9,aspect=0.6);
plt.show()
```

```python
##From pair plot we can conclude that there is much separation in points with plot of year of treatment and auxillary nodes

sns.boxplot(x='year',y='node',data=haber)
plt.show
```

```python
## Violon plot
sns.violinplot(x='year',y='node',data=haber)
plt.show
```
